```python

# Use this cell to begin your analysis, and add as many as you would like!
import matplotlib.pyplot as plt
import pandas as pd

office_episodesDF = pd.read_csv("datasets/office_episodes.csv", parse_dates=["release_date"])
office_episodesDF.head()
```

```python
office_episodesDF.info()
```

```python
colors = []
size = []

for lab, val in office_episodesDF.iterrows():
    if val["scaled_ratings"] < 0.25:
        colors.append("red")
    elif val["scaled_ratings"]  < 0.50:
        colors.append("orange")
    elif val["scaled_ratings"]  < 0.75:
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")
        
    if val["has_guests"] == False:
        size.append(25)
    else:
        size.append(250)
```

```python
office_episodesDF["color"] = colors
office_episodesDF["size"] = size

```

```python
episodes_with_guestsDF = office_episodesDF[office_episodesDF.loc[:,"has_guests"] == True]
episodes_without_guestsDF = office_episodesDF[office_episodesDF.loc[:,"has_guests"] == False]
```

```python
fig = plt.figure()
plt.style.use("fivethirtyeight")
plt.rcParams['figure.figsize'] = [31, 27]

plt.scatter( x = episodes_with_guestsDF.loc[:,"episode_number"], 
             y = episodes_with_guestsDF.loc[:,"viewership_mil"], 
             c = episodes_with_guestsDF.loc[:,"color"], 
             s = episodes_with_guestsDF.loc[:,"size"],
             marker="*"
           )

plt.scatter( x = episodes_without_guestsDF.loc[:,"episode_number"], 
             y = episodes_without_guestsDF.loc[:,"viewership_mil"], 
             c = episodes_without_guestsDF.loc[:,"color"], 
             s = episodes_without_guestsDF.loc[:,"size"]
           )
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.title("Popularity, Quality, and Guest Appearances on the Office")
```
